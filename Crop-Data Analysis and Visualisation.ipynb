```python
# importing Necessary Libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# loading the dataset

crop_data = pd.read_csv("crop_production.csv")
crop_data
```

## Exploratory Data analysis and Data Pre-Processing

```python
# Structure of Data - (Rows and Columns)
crop_data.shape
```

```python
# Columns labels
crop_data.columns
```

```python
# Descriptive Statistics of the data. 
crop_data.describe()
```

```python
# Distict Crops whose data is used in the analysis
print(crop_data['Crop'].unique())
```

```python
crop_data.info()
```

```python
# Checking missing values of the dataset in each column
crop_data.isnull().sum()
```

```python
#Dropping Missing Values
crop_data = crop_data.dropna()
crop_data
```

```python
#Check again for null values left in the dataset. 
crop_data.isnull().values.any()
```

```python
# State Names present in the dataset.
crop_data.State_Name.unique()
```

```python
# Adding a new column "Yield" which indicate Production per unit Area.

crop_data['Yield'] = (crop_data['Production']/crop_data['Area'])
crop_data.head(10)
```

## Data Visualisation

```python
plt.figure(figsize=(10, 6))
plt.bar(crop_data['State_Name'], crop_data['Production'], color='skyblue')
plt.xlabel('State')
plt.ylabel('Total Production')
plt.title('Total Production by State')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=crop_data, x='State_Name', y='Production', palette='viridis')

plt.xticks(rotation = 90)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=crop_data, x='State_Name', y='Yield', palette='viridis')

plt.xticks(rotation = 90)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=crop_data, x='Crop_Year', y='Production', palette='viridis')

plt.xticks(rotation = 90)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=crop_data, x='Crop_Year', y='Yield', palette='viridis')

plt.xticks(rotation = 90)
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=crop_data, x = 'Season', y= 'Yield' , palette='viridis')

plt.xticks(rotation = 90)
```

```python
plt.figure(figsize=(20, 6))
sns.barplot(data = crop_data, x='Crop', y='Yield', palette='viridis')
plt.xticks(rotation = 90)
```

```python
sns.jointplot(data=crop_data, x='Area', y='Production',kind="reg")
plt.xticks(rotation=90)
```

#### Conclusion Obtained from the dataset:
#### 1. Kerala is the largest producer of crop in India
#### 2. production per unit area(yield) of Puducherry is maximunm.
#### 3. In the Year 2011, Crop Yield was maximum.
